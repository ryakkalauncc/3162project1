```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stress_df = pd.read_csv('StressLevelDataset.csv')
stress_df
```

```python
stress_df_clean = stress_df.dropna()
print (stress_df_clean)
```

```python
labels=["Low(0)", "Medium(1)", "High(2)"]
stress_df['stress_level'].value_counts().sort_index().plot(kind='pie'),
autopct = '%1.1f%%',
labels = labels,
title = "Stress Level Distribution"

plt.ylabel("")
plt.show()
```

```python
pd.crosstab(stress_df['academic_performance'],stress_df['stress_level']).plot(kind="bar",figsize=(8,6))
plt.xlabel("Academic Performance")
plt.ylabel("Count")
plt.title("Academic Performance vs Stress Level")
plt.legend(title = "Stress Level", labels=["Low(0)", "Medium(1)", "High(2)"])
plt.show()
```

```python
corr_matrix = stress_df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f",cmap = "coolwarm",cbar=True)
plt.title('Correlation Heatmap of Student Stress')
plt.show()
```

```python
categories = {
    "Psychological":["anxiety_level", "depression","self_esteem","mental_health_history"],
    "Physiological":["sleep_quality","breathing_problem","blood_pressure","headache"],
    "Environmental":["noise_level","living_conditions","safety","basic_needs"],
    "Academic":["academic_performance","study_load","teacher_student_relationship","future_career_concerns"],
    "Social":["social_support","peer_pressure","extracurricular_activities","bullying"]

}
```

```python
category_means={}
for cat, features in categories.items():
  category_means[cat] = stress_df[features].mean(axis=1).corr(stress_df["stress_level"])
category_corr = pd.DataFrame.from_dict(category_means, orient="index",columns=["Correlation with Stress"])
print(category_corr)
```

```python
category_corr.plot(kind="bar",color="skyblue",legend=False,figsize=(8,6))
plt.title("Category Impacts on Stress Level")
plt.ylabel("Correlation with Stress Level")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.lmplot(
    x="sleep_quality",
    y="stress_level",
    data=stress_df,
    height=6,aspect=1.2,
    scatter_kws={"alpha":0.6,"s":60},
    line_kws={"color":"red"}
)
plt.title("Stress Level vs Sleep Quality")
plt.xlabel("Sleep Quality (1=Poor to 5=Excellent)")
plt.ylabel("Stress Level (0=Low, 1=Medium, 2=High)")
plt.show()

```

```python
sns.lmplot(
    x="sleep_quality",
    y="stress_level",
    hue='academic_performance',
    data = stress_df,
    height=6,aspect=1.2,
    scatter_kws={"alpha":0.6,"s":60},
    line_kws={"linewidth":2}
)
plt.title("Stress Level vs Sleep Quality by Academic Performance")
plt.xlabel("Sleep Quality (1=Poor to 5=Excellent)")
plt.ylabel("Stress Level (0=Low, 1=Medium, 2=High)")
plt.show()
```

```python
sns.lmplot(
    x="sleep_quality",
    y="stress_level",
    hue='academic_performance',
    col='academic_performance',
    data=stress_df,
    height=5,aspect=0.9,
    scatter_kws={"alpha":0.6,"s":50},
    line_kws={"linewidth":2}
)
plt.subplots_adjust(top=0.85)
plt.suptitle("Stress Level vs Sleep Quality by Academic Performance",fontsize=14)
plt.show()

```
